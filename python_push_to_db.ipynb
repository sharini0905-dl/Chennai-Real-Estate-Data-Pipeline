```python
# 1. Import libraries
import pandas as pd
import sqlite3
```

```python
# 2. load the csv file into jupyter notebook
df = pd.read_csv('E:\Chennai real estate\clean_data.csv')
df = df.replace('null', None)
df = df.replace('NULL', None)
```

```python
# 3. clean the dataset clean_data.csv
df['bhk'] = pd.to_numeric(df['bhk'], errors = 'coerce').fillna(2)
df['bathroom'] = pd.to_numeric(df['bathroom'], errors = 'coerce').fillna(df['bhk'])
df['age'] = pd.to_numeric(df['age'], errors = 'coerce').fillna(0)
print(df.head(5))
```

```python
# 4. missing values count after cleaning
print("Missing values count after cleaning:\n", df.isnull().sum())
```

```python
# 5. create a SQLite database
conn = sqlite3.connect('real_estate.db')
```

```python
# 6. cleaned data will be directly pushed as a MySQL table 
df.to_sql("chennai_properties", con = conn, if_exists = 'replace', index = False)
```

```python
print("Data has been pushed to the SQLite database successfully.")
conn.close()
```
