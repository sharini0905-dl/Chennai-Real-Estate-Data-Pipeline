```python
#Import packages
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
#1.Dataset creation
data = pd.DataFrame({
    'text':['Payment failed and bill debited', 'Laptop crashed screen blue', 'Need salary payslip leave', 'Office timing cafeteria'],
    'category':['Billing', 'Technical', 'HR', 'General']
})
```

```python
#2.Convert text to vectors for model training
vectorizer = TfidfVectorizer(lowercase = True, stop_words = 'english')
X_train = vectorizer.fit_transform(data['text'])
model = MultinomialNB().fit(X_train, data['category'])
```

```python
#3.New test tickets
test_tickets = [
    'Urgent payment failed twice',
    'wifi not working laptop',
    'where is cafeteria',
    'hello team'
]
```

```python
#4.Predict and print results
urgent_words = ['urgent', 'down', 'not working', 'failed']

for ticket in test_tickets:
    ticket_vector = vectorizer.transform([ticket])
    probs = model.predict_proba(ticket_vector)[0]
    confidence = max(probs) * 100

    category = model.predict(ticket_vector)[0] if confidence >= 30 else 'NEEDS HUMAN REVIEW'
    priority = 'URGENT' if any(w in ticket.lower() for w in urgent_words) else 'NORMAL'

    print(f"Ticket: {ticket} -> Category: {category}| Confidence: {confidence:.1f}% | Priority: {priority}")
```
